```python
from threading import Thread
```

```python
def sum():
  for i in range(1000):
    print("sum hello: ",i)


def sub():
  for i in range(1000):
    print("sub hello: ",i)


thr1 = Thread(target=sum )
thr2 = Thread(target=sub )

thr1.start()
thr1.join()
thr2.start()
print("hello thread")
```

```python
#dynamic args in python

def single_arg(*ar):
  for i in ar:
    print(i)

single_arg(1,2)
single_arg(1,2,3,4,5) 

```

```python
def single_arg(**ar):
  for i in ar:
    print(i)

single_arg(key1=1,key2=2)
single_arg(key1=1,key2=2,key3=3,key4=4,key5=5) 
```
